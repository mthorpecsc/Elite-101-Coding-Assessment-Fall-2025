## Exercise 1: Create a DataFrame from a Dictionary

Dictionaries map column names to lists of values — a natural way to build a DataFrame.

```python
# Exercise 1: Create a DataFrame from a dictionary

#   'name': ['Alice', 'Bob', 'Charlie', 'Diana']
#   'age': [25, 30, 35, 28]
#   'department': ['Engineering', 'Marketing', 'Engineering', 'Sales']
#   'salary': [75000, 65000, 80000, 70000]

data = {
    'name':  ['Alice', 'Bob', 'Charlie', 'Diana'],
    'age': [25, 30, 35, 28],
    'department': ['Engineering', 'Marketing', 'Engineering', 'Sales'],
    'salary': [75000, 65000, 80000, 70000]

}
my_df = pd.DataFrame(data)
print(my_df)
```

## Exercise 3: `.loc[]` vs `.iloc[]`

- `.loc[]` — selects by **label** (column name, row label)
- `.iloc[]` — selects by **integer position** (row number, column number)

```python
# Exercise 3: Practice .loc[] and .iloc[]
# Using the employee DataFrame loaded at the top:
print(df.head())
print()

# Task 1: Use .iloc[] to get the first 3 rows
# TODO: your code here
first_three = df[0:3]
print("First 3 rows (iloc):")
print(first_three)
print()

# Task 2: Use .loc[] to get all rows where department is 'Engineering'
# Hint: .loc[condition]
# TODO: your code here
engineering = df.loc[df['department'] == 'Engineering']
print("Engineering department (loc):")
print(engineering)
print()

# Task 3: Use .iloc[] to get row 2, columns 0 through 2
# TODO: your code here
subset = df.iloc[2,0:3]
print("Row 2, first 3 columns (iloc):")
print(subset)

```
